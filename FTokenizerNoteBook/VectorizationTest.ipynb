```python
import tensorflow as tf
import numpy as np
import pandas as pd
import json
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
```

```python
with open("E:/MLCourse/TF2/Notebook/tutorialNotebook/data/data (1).json","r",encoding="utf-8") as f:
    data=json.load(f)
data=data["bạn bè"]["câu hỏi"]
data
```

```python
vectorize_layer=TextVectorization(
    max_tokens=20000,
    standardize="lower_and_strip_punctuation",
    output_mode='int',
    output_sequence_length=40,
    split='whitespace',
    ngrams=2
)
vectorize_layer.adapt(data)
```

```python
vectorize_layer(["bạn thân của bạn là ai"])
```
